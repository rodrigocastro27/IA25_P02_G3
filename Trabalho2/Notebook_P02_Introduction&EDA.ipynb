### Notebook 3 : Regras de Associação com Apriori

A Fórmula 1 produz uma enorme quantidade de dados relacionados com corridas, equipas, pilotos, circuitos e resultados. A análise de regras de associação será aplicada para explorar a coocorrência de pilotos nos lugares do pódio. Este estudo é essencial para compreender:

1. Rivalidades marcantes entre pilotos;

2. Combinações frequentes de pilotos que tendem a estar juntos no pódio;

4. Padrões característicos como por exemplo, anos de domínio da Ferrari, Red Bull ou Mercedes.

<br>

O Notebook 3 tem como objetivo implementar o algoritmo Apriori para descobrir e quantificar regras de associação entre pilotos, onde o foco será transformar o resultado das corridas numa perspetiva de avaliação de balanço de mercado, onde cada corrida é tratada como uma transação e cada piloto que termina no pódio é um item da mesma.  
<br>

Regras de associação que revelam relações como:
> **“Se o piloto X aparece no pódio, o piloto Y tende a aparecer também”**

<br>
Arelevância das regras de associação descobertas pelo algoritmo Apriori são quantificadas através de três métricas-chave:

- Confidence: Indica a probabilidade condicional de o Piloto Y estar no pódio, dado que o Piloto X já lá está.

- Lift: Medida de interesse da regra. Compara a Confiança da regra com a frequência esperada de Y (Piloto Y) no pódio. Valores de Lift superiores a 1 indicam que a coocorrência é mais frequente do que seria por acaso.

<br>
Este estudo permite compreender melhor o panorama competitivo da Fórmula 1 ao longo das décadas, identificar rivalidades históricas e revelar padrões que não são imediatamente visíveis apenas pelos resultados brutos.

**Debut Decade of Drivers in F1 History** - Modelling and Univariate Analysis

```python
# Feature Engineering: Debut Decade of Drivers in F1 History
# Join datasets to find the debut year of each driver
participation = pd.merge(
    results[['raceId', 'driverId']],
    races[['raceId', 'year']],
    on='raceId',
    how='left'
)

# Find the debut year for each driver
debut_year = participation.groupby('driverId')['year'].min().reset_index()
debut_year.rename(columns={'year': 'Debut_Year'}, inplace=True)

# Calculate Debut Decade
debut_year['Debut_Decade'] = (
    (debut_year['Debut_Year'] // 10 * 10).astype(str) + 's'
)

# Count drivers by debut decade
debut_counts = debut_year['Debut_Decade'].value_counts().sort_index()
```

 **Total Points Available for each Decade** -  Modelling and Univariate Analysis

```python
# Join results and races to get points per race
points_per_race = pd.merge(
    results[['raceId', 'points']],
    races[['raceId', 'year']],
    on='raceId',
    how='left'
)

# Calculate points per race per each year 
points_per_race['Race_Decade'] = (
    (points_per_race['year'] // 10 * 10).astype(str) + 's'
)

# Group by
total_points_by_decade = points_per_race.groupby('Race_Decade')['points'].sum().sort_index()
total_points_by_decade = total_points_by_decade.dropna()

print("\n--- Table: Total Points Available per Decade ---")
print(total_points_by_decade.to_frame(name='Total Points'))

plt.figure(figsize = (14, 7))
plt.bar(total_points_by_decade.index, total_points_by_decade.values, color='gold', edgecolor='black', alpha=0.9)
plt.xticks(rotation=45, ha='right')
plt.ylabel("Total Points ")
plt.xlabel("Race Decade")
plt.title("Points Available per Decade in F1 History")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

##### Bivariate Analysis

```python
# Calculate Total Career Points for each driver
total_points = results.groupby('driverId')['points'].sum().reset_index()
total_points.rename(columns={'points': 'Total_Career_Points'}, inplace=True)

# Calculate Ammount of Races Started for each driver
races_started = results.groupby('driverId').size().reset_index(name='Races_Started')

# join total points and races started
performance = pd.merge(total_points, races_started, on='driverId')

# Average Points per Race
performance['Avg_Points_per_Race'] = (
    performance['Total_Career_Points'] / performance['Races_Started']
)


bivariate = pd.merge(
    debut_year[['driverId', 'Debut_Decade']],
    performance[['driverId', 'Avg_Points_per_Race']], 
    on='driverId',
    how='inner'
)

# Group by Debut Decade and calculate mean Avg Points
normalized_performance = (
    bivariate.groupby('Debut_Decade')['Avg_Points_per_Race']
                           .mean()
                           .sort_index()
)
normalized_performance = normalized_performance.dropna()

print("\n--- Average Performance per Debut Decade in F1 History ---")
print(normalized_performance.to_frame(name='Averege Points/Race'))


# 3. VISUALIZAÇÃO
plt.figure(figsize = (14, 7))
plt.bar(normalized_performance.index, normalized_performance.values, color='teal', alpha=0.9)
plt.xticks(rotation=45, ha='right')
plt.ylabel("Average Points per Race")
plt.xlabel("Debut Decade")
plt.title("Average Driver Points per Race per Debut Decade in F1 History")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

##### Outliers Detection IQR

```python
def detect_outliers(df,features):
    outlier_indices = []

    for c in features:
        # 1st quartile
        Q1 = np.percentile(df[c],25)
        # 3rd quartile
        Q3 = np.percentile(df[c],75)
        # IQR
        IQR = Q3 - Q1
        # Outlier step
        outlier_step = IQR * 1.5
        # detect outlier and their indeces
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        # store indeces
        outlier_indices.extend(outlier_list_col)


    outlier_indices = Counter(outlier_indices)
    return list(outlier_indices.keys())

# Outliers Detection using IQR method
performance_cleaned = performance.dropna(subset=['Avg_Points_per_Race']).reset_index(drop=True)
outlier_indices = detect_outliers(performance_cleaned, ['Avg_Points_per_Race'])

# Extract the DataFrames of Outliers
outliers_df = performance_cleaned.loc[outlier_indices].sort_values(
    by='Avg_Points_per_Race', ascending=False
)

# Calculate statistics for IQR output
Q1 = performance_cleaned['Avg_Points_per_Race'].quantile(0.25)
Q3 = performance_cleaned['Avg_Points_per_Race'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR


plt.figure(figsize = (10, 6))
plt.boxplot(performance_cleaned['Avg_Points_per_Race'], vert=False)
plt.xlabel("Averafe Points per Race")
plt.title("Box Plot: Performance of Drivers in F1 History (Avg Points per Race")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show() 


print(f"Number of Drivers Outliers (Performance): {len(outliers_df)}")

# Join with drivers to get names
outliers_named = pd.merge(outliers_df, drivers[['driverId', 'forename', 'surname']], on='driverId', how='left')
outliers_named['Driver'] = outliers_named['forename'] + ' ' + outliers_named['surname']
print("\n--- Top 10 Drivers Outliers (Most Efficient in Points/Race) ---")
print(outliers_named[['Driver', 'Avg_Points_per_Race']].head(10).to_string(index=False))
```

```python
# Points Outlier Detection using IQR method
points_data = driver_standings['points'].dropna()

# -----------------------------------------------
# Visualization

plt.figure(figsize=(10, 6))
plt.boxplot(points_data)
plt.title("Boxplot for Points in Driver Standings")
plt.ylabel("Driver Standings")
plt.show()



def detect_outliers_iqr_dstand(data):
    
    # 1st quartile
    Q1 = np.percentile(data, 25)
    # 3rd quartile
    Q3 = np.percentile(data, 75)
    # IQR
    IQR = Q3 - Q1
    # Outlier step (1.5 * IQR)
    outlier_step = IQR * 1.5

    # detect outlier and their indeces
    outlier_values = data[(data < Q1 - outlier_step) | (data > Q3 + outlier_step)]

    return outlier_values

# Execute the outlier detection
outliers_in_points = detect_outliers_iqr_dstand(points_data)

print("\n--- Results ---")
print(f"Total Amount of Outliers Detected: {len(outliers_in_points)}")
print(f"This corresponds to {len(outliers_in_points)/driver_standings.shape[0]:.2%} of all DATA.")

if not outliers_in_points.empty:
    print("\n5 biggest points accumulated:")
    print(outliers_in_points.nlargest(5))
else:
    print("No outlier detected.")
```

##### Feature Engineering

**Get the amount of DNFs / DNS/ DSQ in a F1 Season**

```python
# Classify the failure reasons into broader categories
technical_accident_keywords = ['Engine', 'Transmission', 'Accident', 'Suspension', 'Gearbox', 'Collision', 'Brakes', 'Hydraulics', 'Ignition']
dns_dsq_keywords = ['Disqualified', 'Did not start', 'Excluded', 'Withdrew', 'Injury', 'Retired'] 

def classify_failure(status_str):
    if status_str == 'Finished':
        return 'Finished'
    elif pd.isna(status_str):
         return 'Status Indefinido'
    elif any(keyword in status_str for keyword in technical_accident_keywords):
        return 'Technical/Accident'
    elif any(keyword in status_str for keyword in dns_dsq_keywords):
        return 'DNS/DSQ/Admin'
    else:
        return 'Other DNF'

# Apply the classification function to create a new column
results_status['Failure_Category'] = results_status['status'].apply(classify_failure)

# Create a bivariate DataFrame with Failure
bivariate = pd.merge(
    results_status[['raceId', 'Failure_Category']],
    races[['raceId', 'year']],
    on='raceId',
    how='left'
)

# Criamos a feature da Década
bivariate['Decade'] = (bivariate['year'] // 10 * 10).astype(str) + 's'
```

##### Bivariate Analysis

**Evolution of Race Result Categories per Decade in F1 History**

```python
# Group by Decade and Failure_Category to get counts
category_per_decade = bivariate.groupby(['Decade', 'Failure_Category']).size().unstack(fill_value=0)

# calculate proportions
total_per_decade = category_per_decade.sum(axis=1)
proportion_per_decade = category_per_decade.divide(total_per_decade, axis=0) * 100

print("\n--- Proporção de Status por Década (%) ---")
print(proportion_per_decade.round(1))



# order the columns
order = ['Finished', 'Technical/Accident', 'DNS/DSQ/Admin', 'Other DNF', 'Status Unknown']
proportion_per_decade = proportion_per_decade.reindex(columns=order, fill_value=0)


colors = ['#4CAF50', '#FF9800', '#2196F3', '#F44336', '#9E9E9E'] 

plt.figure(figsize=(15, 8))
proportion_per_decade.plot(kind='bar', stacked=True, color=colors, ax=plt.gca())
plt.xlabel("Decade")
plt.ylabel("Propotion of Race finishing status per decade (%)")
plt.title("Evolution of Race Result Categories per Decade in F1 History")
plt.xticks(rotation=45, ha='right')
plt.legend(title='Results Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show() # 
```
